## Training model

```python
from sklearn.model_selection import train_test_split
```

```python
y = aligned.Class
X = aligned.drop('Class', axis=1)
train_ratio = 0.7
test_ratio = 0.20
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=train_ratio, random_state=1)
```

```python
from sklearn import svm
```

```python
svm_classifier = svm.SVC()
```

```python
svm_classifier.fit(X_train, y_train)
```

## Measuring performance

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, f1_score
```

```python
print(confusion_matrix(y_test, svm_classifier.predict(X_test)))
```

```python
print(accuracy_score(y_test, svm_classifier.predict(X_test)))
```
